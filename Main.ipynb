```python
# Import libraries
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

```

```python
df_train = pd.read_csv('formulatedtrain.csv')
df_test = pd.read_csv('formulatedtest.csv')
df_original_test = pd.read_csv('test.csv')
```

```python
X_tra=df_train.drop(['SalePrice'],axis=1)
Y_tra=df_train['SalePrice']

x_test=df_test
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_tra, Y_tra, test_size=0.3, random_state=44)
model = linear_model.LinearRegression()
model.fit(X_train, Y_train)
y_pred_train = model.predict(X_train)

print("\nTrain Test: ", y_pred_train[0:10])
```

# Modelin RMSE vsç gibi doğruluk ölçütlerinin hesaplanması

```python
def rmse_cv(model):
    rmse = np.sqrt(-cross_val_score(model, X_train, Y_train, scoring="neg_mean_squared_error", cv=5)).mean()
    return rmse
    
```

```python
import math 
MSE = np.square(np.subtract(Y_train,y_pred_train)).mean() 
RMSE = math.sqrt(MSE)

RMSE_Cross_Val = rmse_cv(model)

R2_Score =  r2_score(Y_train, y_pred_train)

```
